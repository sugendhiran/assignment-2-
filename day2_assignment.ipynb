```python
x=[1,2,3,4]
y=[11,21,31,12]
x1=[4,3,7,8]
y1=[12,23,33,15]
plt.bar(x,y,label="x y variation",color='g')
plt.bar(x1,y1,label="x1 y1 variation")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```
